```python
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# Sample Data set
X = np.array([[1], [2], [3], [4], [5]])   # Features
y = np.array([5, 7, 9, 11, 13])           # Target (y = 2x + 3)
```

```python
# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Create Model with desired learning rate
model = SGDRegressor(learning_rate='constant', eta0=0.01, max_iter=1000, tol=1e-3,random_state=42)
```

```python
# Train the model
model.fit(X_scaled, y)
```

```python
# Predictions
predictions = model.predict(X_scaled)
```

```python
# Evaluate the model
mse = mean_squared_error(y, predictions)
print(f"Mean Squared Error: {mse}")

print("Weights (Coefficients):", model.coef_)
print("Bias (Intercept):", model.intercept_)

```

```python
# Plot
import matplotlib.pyplot as plt

plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, predictions, color='red', label='Predicted')
plt.legend()
plt.title("Linear Regression via SGD (Gradient Descent)")
plt.xlabel("X")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
